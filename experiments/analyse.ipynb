```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Chargement de fichier 
df = pd.read_csv('results/resultHypo1.csv', sep=';')
```

```python
print(df.head())
```

```python
victoire = df[df['Result'] == 1]
null = df[df['Result'] == 0]
defaite = df[df['Result'] == -1]
# Result Nodes NbGames Depth sizeGrid
```

```python
# Affichage de la courbe d'évolution du nombre de victoire, nul et défaite
# Calcul de la moyenne de victoire pour chaque profondeur
nbVictoireParProfondeur = victoire.groupby('Depth')['Result'].count()
nbNullParProfondeur = null.groupby('Depth')['Result'].count()
nbDefaiteParProfondeur = defaite.groupby('Depth')['Result'].count()

#Proportion de victoire par profondeur
propVictoireParProfondeur = nbVictoireParProfondeur + nbNullParProfondeur / (nbVictoireParProfondeur + nbNullParProfondeur + nbDefaiteParProfondeur)

# Affichage de la courbe d'évolution du nombre de victoire
fig, ax = plt.subplots()
ax.plot(propVictoireParProfondeur, 'o-')
ax.set(xlabel='Profondeur', ylabel='Proportion de victoire',
         title='Proportion de victoire en fonction de la profondeur')
ax.grid()
plt.show()
fig.savefig('propVictoireParProfondeur.png')


```
